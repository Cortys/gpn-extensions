```python
import numpy as np
import scipy.special as ss
from scipy.stats import dirichlet
import matplotlib.pyplot as plt

ss.beta(5.5, 5.5)
```

```python
np.sqrt(ss.beta(1, 10) * ss.beta(10, 1))
```

```python
x = np.linspace(0, 1, 1000)
y = 1 - x
ps = np.concatenate([x[:, None], y[:, None]], -1)

a = 100
b = 1
f = 0.5

y1 = np.array([dirichlet.pdf(p, [f * a + (1-f) * b, (1-f) * a + f * b]) for p in ps])
y2 = np.array([np.power(dirichlet.pdf(p, [a, b]), f) * np.power(dirichlet.pdf(p, [b, a]), 1-f) for p in ps])
y2_scale = np.sum(y2) / len(x)
y2 = y2 / y2_scale

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```
